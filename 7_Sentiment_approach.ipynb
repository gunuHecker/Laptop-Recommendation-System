```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('sentence-transformers/all-mpnet-base-v2')
```

```python
data_p = '/content/gdrive/MyDrive/rtings-laptops-merge.csv'
data = pd.read_csv( data_p)
data.head()
```

```python
def create_large_sentence(row):
    sentence = f"{row['Gaming Review'].strip()} Its Gaming Rating is {row['Gaming Rating']}. " \
               f"{row['Multimedia Review'].strip()} Its Multimedia Rating is {row['Multimedia Rating']}. " \
               f"{row['Business Review'].strip()} Its Business Rating is {row['Business Rating']}."
    return sentence
```

```python
data['Large Sentence'] = data.apply(create_large_sentence, axis=1)
```

```python
print(data[['Large Sentence']].head())
```

```python
def calculate_embedding(sentence):
    return model.encode(sentence)
```

```python
data['Embedding'] = data['Large Sentence'].apply(calculate_embedding)
```

```python
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
query_sentence = "Maximum gaming rating and maximum multimedia rating and maximum business rating."

query_embedding = model.encode(query_sentence)
```

```python
data['Similarity'] = data['Embedding'].apply(lambda x: cosine_similarity(query_embedding, x))

top_5_models = data.sort_values(by='Similarity', ascending=False).head(5)
```

```python
pd.set_option('display.width', 1000)
print(top_5_models[['Laptop Name', 'Gaming Rating' , 'Multimedia Rating' , 'Business Rating' ,'Similarity', 'Refresh Rate', 'RAM', 'Storage', 'Battery','CPU']])
```
